## Processing hyperparameters from text format

```python
import pandas as pd
import numpy as np
_FOLDER_2 = "results/"
```

```python
# Coefficient 1
coef=1
best_parameters=  ["DRUG_ID	best_alpha3	best_gamma3	best_coef03	best_degree3",
"170	100.0	0.10000	5.0	1.0",
"173	100.0	0.01000	5.0	3.0",
"180	100.0	0.00001	0.5	1.0",
"200	100.0	0.00001	5.0	1.0",
"219	10.0	0.00001	0.5	2.0",
"272	500.0	0.00010	5.0	2.0",
"273	0.5	0.01000	1.0	1.0",
"274	1.0	0.00001	0.5	5.0",
"276	1.0	0.00001	5.0	5.0",
"328	5.0	0.00010	0.5	3.0",
"346	1.0	0.01000	0.5	1.0"]

param_dict = {}
parameters_names = best_parameters[0].split("\t")
for i in range(1, len(best_parameters)):
    for j,param in enumerate(parameters_names):
        if param in param_dict:
            param_dict[param].append(best_parameters[i].split("\t")[j])
        else:
            param_dict[param] = [best_parameters[i].split("\t")[j]]
            
hyperparameters = pd.DataFrame(param_dict)
hyperparameters["DRUG_ID"] = np.int32(hyperparameters["DRUG_ID"])
hyperparameters.set_index("DRUG_ID", inplace= True)
for col in hyperparameters.columns:
    hyperparameters[col] = np.float32(hyperparameters[col])

hyperparameters.to_csv(_FOLDER_2+"hyperparameters_coef"+str(coef)+"_alg1.csv")    
hyperparameters
```
